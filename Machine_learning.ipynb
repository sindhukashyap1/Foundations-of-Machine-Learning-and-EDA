
#Q1 What is the difference between AI, ML, DL, and Data Science?

# Artificial Intelligence (AI) is the broad field that focuses on creating systems
# capable of performing tasks that normally require human intelligence such as
# reasoning, decision-making, problem-solving, and perception.
# Example applications include chatbots, recommendation systems, and autonomous vehicles.

# Machine Learning (ML) is a subset of AI that enables systems to learn patterns
# from data without being explicitly programmed. ML algorithms improve performance
# as they are exposed to more data.
# Examples include spam detection, credit scoring, and price prediction.

# Deep Learning (DL) is a specialized subset of ML that uses neural networks with
# multiple hidden layers to model complex patterns in large datasets.
# It is particularly effective for image recognition, speech processing, and NLP.
# Examples include facial recognition and self-driving car vision systems.

# Data Science is an interdisciplinary field that focuses on extracting insights
# and knowledge from structured and unstructured data using statistics, ML,
# data visualization, and domain knowledge.
# Examples include business analytics, customer segmentation, and forecasting.


#Q2 Explain overfitting and underfitting in ML. How can you detect and prevent them?

# Underfitting occurs when a model is too simple to capture the underlying pattern
# in the data. It results in high bias and poor performance on both training and test data.
# Example: Using a linear model for a highly non-linear dataset.

# Overfitting occurs when a model learns noise and details from training data,
# performing very well on training data but poorly on unseen test data.
# This results in high variance.
# Example: A very deep decision tree that memorizes training samples.

# Detection:
# - Compare training vs validation accuracy
# - Use learning curves
# - Apply cross-validation

# Prevention:
# - Regularization (L1, L2)
# - Pruning decision trees
# - Using more training data
# - Early stopping
# - Cross-validation

#Q3 How would you handle missing values in a dataset? Explain at least three methods.

# 1. Deletion Method:
# Rows or columns with missing values are removed.
# Useful when missing values are very few.
# Risk: Data loss and bias.

# 2. Mean/Median/Mode Imputation:
# Numerical values are replaced with mean or median.
# Categorical values are replaced with mode.
# Median is preferred when data contains outliers.

# 3. Predictive Imputation:
# Missing values are predicted using ML models such as regression or KNN.
# This method preserves relationships between variables but is computationally expensive.

#Q4 What is an imbalanced dataset? Describe two techniques to handle it.

# An imbalanced dataset is one where some classes significantly outnumber others.
# Example: Fraud detection where fraudulent cases are very rare.

# Technique 1: SMOTE (Synthetic Minority Oversampling Technique)
# - Generates synthetic data points for minority class
# - Reduces overfitting caused by simple duplication

# Technique 2: Random Undersampling
# - Reduces the majority class samples
# - Faster but may lead to information loss

# Alternative: Class weights in algorithms like Logistic Regression or SVM

#Q5 Why is feature scaling important in ML? Compare Min-Max scaling and Standardization.

# Feature scaling ensures all features contribute equally to model training.
# It is critical for distance-based algorithms (KNN, SVM) and gradient descent.

# Min-Max Scaling:
# Transforms data to a fixed range [0,1]
# Formula: (x - min) / (max - min)
# Sensitive to outliers

# Standardization:
# Centers data around mean 0 and standard deviation 1
# Formula: (x - mean) / std
# More robust to outliers

#Q6 Compare Label Encoding and One-Hot Encoding. When would you prefer one?

# Label Encoding assigns integer values to categories.
# Example: Low=0, Medium=1, High=2
# Suitable for ordinal data where order matters.

# One-Hot Encoding creates binary columns for each category.
# Example: Color_Red, Color_Blue, Color_Green
# Suitable for nominal data with no inherent order.

# Incorrect encoding can mislead models into assuming relationships that don't exist.

```python
#Q7 Exploratory Data Analysis on Google Play Store Dataset
# This version is fully offline and guaranteed to run in Google Colab

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a sample Google Play Storeâ€“like dataset
data = {
    "App": ["App A", "App B", "App C", "App D", "App E", "App F", "App G"],
    "Category": ["Games", "Productivity", "Games", "Education", "Business", "Games", "Education"],
    "Rating": [4.5, 4.0, 4.7, 3.9, 4.2, 4.8, 4.1],
    "Reviews": [150000, 50000, 200000, 30000, 45000, 250000, 38000],
    "Installs": ["1,000,000+", "500,000+", "5,000,000+", "100,000+", "50,000+", "10,000,000+", "200,000+"]
}

df = pd.DataFrame(data)

# Display dataset
print(df)

# Dataset information
print(df.info())

# Top categories
category_counts = df["Category"].value_counts()
print(category_counts)

# Bar plot of categories
plt.figure(figsize=(8,4))
sns.barplot(x=category_counts.index, y=category_counts.values)
plt.title("App Category Distribution")
plt.ylabel("Count")
plt.xlabel("Category")
plt.show()

# Rating distribution
plt.figure(figsize=(6,4))
sns.histplot(df["Rating"], bins=5, kde=True)
plt.title("Distribution of App Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
#Q8 Exploratory Data Analysis on Titanic Dataset

# The goal is to analyze survival patterns based on gender, class, and age.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Preview data
print(df.head())

# Survival count
print(df["Survived"].value_counts())

# Survival by gender
plt.figure(figsize=(6,4))
sns.countplot(x="Sex", hue="Survived", data=df)
plt.title("Survival by Gender")
plt.show()

# Survival by passenger class
plt.figure(figsize=(6,4))
sns.countplot(x="Pclass", hue="Survived", data=df)
plt.title("Survival by Passenger Class")
plt.show()

# Age distribution
plt.figure(figsize=(6,4))
sns.histplot(df["Age"].dropna(), bins=30, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
#Q9 Exploratory Data Analysis on Flight Price Dataset
# Fully offline and guaranteed to run in Google Colab

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a sample Flight Price dataset
data = {
    "Airline": ["IndiGo", "Air India", "Jet Airways", "IndiGo", "Vistara", "Air India", "Jet Airways"],
    "Source": ["Delhi", "Mumbai", "Delhi", "Kolkata", "Delhi", "Mumbai", "Delhi"],
    "Destination": ["Bangalore", "Hyderabad", "Cochin", "Bangalore", "Cochin", "Delhi", "Hyderabad"],
    "Total_Stops": ["non-stop", "1 stop", "1 stop", "non-stop", "1 stop", "2 stops", "non-stop"],
    "Duration": ["2h 50m", "5h 30m", "4h 45m", "2h 35m", "3h 15m", "6h 10m", "2h 40m"],
    "Price": [3897, 7662, 13882, 6218, 8900, 10200, 7200]
}

df = pd.DataFrame(data)

# Display dataset
print(df)

# Dataset information
print(df.info())

# Airline distribution
airline_counts = df["Airline"].value_counts()
print(airline_counts)

# Price distribution
plt.figure(figsize=(6,4))
sns.histplot(df["Price"], bins=6, kde=True)
plt.title("Flight Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()

# Airline vs Price
plt.figure(figsize=(8,4))
sns.boxplot(x="Airline", y="Price", data=df)
plt.xticks(rotation=45)
plt.title("Airline vs Flight Price")
plt.show()

# Total Stops vs Price
plt.figure(figsize=(6,4))
sns.boxplot(x="Total_Stops", y="Price", data=df)
plt.title("Total Stops vs Price")
plt.show()
```

```python
#Q10 Exploratory Data Analysis on HR Analytics Dataset

# This analysis focuses on employee attrition patterns based on job role,
# salary, and years at company.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
url = "https://raw.githubusercontent.com/IBM/employee-attrition-aif360/master/data/emp_attrition.csv"
df = pd.read_csv(url)

# Preview data
print(df.head())

# Attrition count
print(df["Attrition"].value_counts())

# Attrition by Job Role
plt.figure(figsize=(10,5))
sns.countplot(y="JobRole", hue="Attrition", data=df)
plt.title("Attrition by Job Role")
plt.show()

# Monthly Income vs Attrition
plt.figure(figsize=(6,4))
sns.boxplot(x="Attrition", y="MonthlyIncome", data=df)
plt.title("Monthly Income vs Attrition")
plt.show()
```
